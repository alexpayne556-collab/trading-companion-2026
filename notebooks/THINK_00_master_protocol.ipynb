## PROTOCOL 1: KILL THE OBVIOUS

````python
obvious_answer = """
[Write the answer any AI would give]
[The common wisdom]
[What appears in every course/blog/video]
"""

print("THE OBVIOUS ANSWER (to be deleted):")
print(obvious_answer)
print("\nâŒ THIS IS WHAT EVERYONE KNOWS. NO EDGE HERE. DELETED.")
````

## PROTOCOL 2: EXPOSE ASSUMPTIONS

What does the question ASSUME to be true?
Run THINK_01_assumption_destroyer.ipynb for deep version.

````python
assumptions = [
    "1. [What does this question assume?]",
    "2. [What else?]",
    "3. [What's taken for granted?]",
    "4. [What if that's wrong?]",
    "5. [What's never questioned?]",
]

print("ASSUMPTIONS IN THE QUESTION:")
for a in assumptions:
    print(a)
````

## PROTOCOL 3: MAP THE PLAYERS

Who is involved? Who profits? Who loses?
Run THINK_02_player_mapper.ipynb for deep version.

````python
players = {
    "Visible Players": [
        "[Who you think about first]",
        "[The obvious participants]",
    ],
    "Invisible Players": [
        "[Who profits but you don't see?]",
        "[Who is one level above the obvious?]",
        "[Who controls timing/information?]",
    ],
    "Where Are We?": "[Which player are we? What position?]",
}

print("PLAYER MAP:")
for category, items in players.items():
    print(f"\n{category}:")
    if isinstance(items, list):
        for item in items:
            print(f"  - {item}")
    else:
        print(f"  {items}")
````

## PROTOCOL 5: INVERT THE QUESTION

Flip the question completely. Work backward from the inversion.

````python
inversion = """
ORIGINAL: [The question as asked]

INVERTED: [The opposite question]
- If asked "what will succeed?" â†’ "what will FAIL and why?"
- If asked "what's the opportunity?" â†’ "what's the TRAP?"
- If asked "how do we win?" â†’ "how do we definitely LOSE?"

INVERTED ANSWER:
[Answer the inverted question]

WORKING BACKWARD:
[What does avoiding the inversion tell us about the original?]
"""

print("INVERSION:")
print(inversion)
````
