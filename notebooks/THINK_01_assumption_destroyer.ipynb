```python
# THE QUESTION TO ANALYZE
# Change this to whatever we're thinking about

QUESTION = """
How do we predict which stocks will run?
"""
```

```python
# STEP 1: EXTRACT THE ASSUMPTIONS IN THE QUESTION ITSELF
# What does this question ASSUME to be true?

assumptions_in_question = [
    "Stocks 'run' in a predictable way",
    "Prediction is possible with available information",
    "The goal is to predict (vs avoid, vs react, vs something else)",
    "'We' can act on predictions fast enough",
    "The prediction would still work if others knew it",
]

print("ASSUMPTIONS HIDDEN IN THE QUESTION:")
print("=" * 50)
for i, a in enumerate(assumptions_in_question, 1):
    print(f"{i}. {a}")
```

```python
# STEP 2: WHAT IF EACH ASSUMPTION IS WRONG?

inversions = {
    "Stocks 'run' in a predictable way": 
        "What if runs are manufactured events, not natural patterns?",
    
    "Prediction is possible with available information":
        "What if the information that matters is deliberately hidden?",
    
    "The goal is to predict":
        "What if the better goal is to NOT BE THE EXIT LIQUIDITY?",
    
    "'We' can act on predictions fast enough":
        "What if by the time we see it, we're already too late?",
    
    "The prediction would still work if others knew it":
        "What if any public edge gets arbitraged away instantly?",
}

print("IF EACH ASSUMPTION IS WRONG:")
print("=" * 50)
for assumption, inversion in inversions.items():
    print(f"\nASSUMPTION: {assumption}")
    print(f"INVERSION:  {inversion}")
```

```python
# STEP 3: WHAT'S THE BETTER QUESTION?
# Based on the inversions, what should we ACTUALLY be asking?

better_questions = [
    "How do we avoid being exit liquidity for someone else's position?",
    "What information is being hidden and by whom?",
    "When is a 'setup' actually a trap?",
    "What game are we actually playing, and who are the other players?",
    "What would still work even if everyone knew about it?",
]

print("BETTER QUESTIONS TO ASK:")
print("=" * 50)
for i, q in enumerate(better_questions, 1):
    print(f"{i}. {q}")
```

```python
# STEP 4: WHAT DOES THE OBVIOUS ANSWER ASSUME?
# If I just answered the original question, what would I say?
# And what does THAT assume?

obvious_answer = """
Use technical indicators (RSI, MACD, volume), 
find catalysts (news, earnings), 
look for sector momentum, 
manage risk with stop losses.
"""

obvious_answer_assumes = [
    "Technical indicators reveal real information (vs noise)",
    "Catalysts cause moves (vs moves cause catalysts to be released)",
    "Sector momentum is natural (vs manufactured)",
    "Stop losses protect you (vs they get hunted)",
    "This advice isn't already priced in",
    "This is different from what everyone else does (it's not)",
]

print("THE OBVIOUS ANSWER:")
print(obvious_answer)
print("\nWHAT THE OBVIOUS ANSWER ASSUMES:")
print("=" * 50)
for i, a in enumerate(obvious_answer_assumes, 1):
    print(f"{i}. {a}")
```

```python
# STEP 5: THE LEVEL 5 REFRAME
# Now that we've destroyed assumptions, what's the ACTUAL insight?

level_5_insight = """
ORIGINAL QUESTION: How do we predict which stocks will run?

LEVEL 5 REFRAME: 
The question assumes prediction is the game. It's not.

The game is: Who is on the other side of my trade?
- If I'm buying from someone who KNOWS more, I lose.
- If I'm buying from someone who NEEDS to exit, I might win.

The question isn't "what will run?"
The question is "when am I NOT the sucker at the table?"

ACTIONABLE REFRAME:
Instead of predicting winners, identify situations where
the person selling to me is forced to sell (not choosing to).
- Short squeezes (shorts MUST cover)
- Index rebalancing (funds MUST buy)
- Tax loss selling ending (forced selling stops)
- Margin calls (forced liquidation)

These are situations where I'm not competing against
someone with more information. I'm trading against
someone with LESS CHOICE.
"""

print(level_5_insight)
```
