# PART 1: ORDER FLOW - What Happens When You Hit "Buy"

## What I THINK I know:

1. You submit order through broker (Robinhood, Schwab, etc.)
2. Broker has choices:
   - Route to exchange (NYSE, NASDAQ)
   - Route to market maker (Citadel, Virtu)
   - Internalize (fill from own inventory)
3. Most retail orders go to market makers via PFOF (Payment for Order Flow)
4. Market maker sees your order BEFORE filling it
5. They can trade against you, with you, or just take the spread

# PART 3: OPTIONS MECHANICS - How Options Move Stock

## What I THINK I know:

1. Options = contracts to buy (call) or sell (put) at a strike price
2. Options market makers sell options and must HEDGE
3. Hedging a call = buying stock (delta hedging)
4. As stock approaches strike, delta increases = more hedging = more buying
5. This is "gamma" - the acceleration of delta
6. Gamma squeeze = MM forced to buy more and more as price rises
7. Max pain = price where most options expire worthless (MM keeps premium)

## What I DON'T know:

- How QUICKLY do MMs hedge? Every tick? End of day?
- What's the threshold for "significant" options activity?
- How do we SEE options positioning in real-time?
- When does gamma exposure actually FORCE the move?
- How accurate is "max pain" as a predictor?
- What's the relationship between options expiration and stock moves?

## Questions for Fenrir:

4. What time do options expire? How does that affect stock?

# PART 8: THE PLAYERS - Who's Actually In The Game

## The Hierarchy (as I understand it):

### Tier 1: The Infrastructure
- Exchanges (NYSE, NASDAQ, CBOE)
- Clearinghouse (DTCC)
- Regulators (SEC, FINRA)
- Set the rules. Profit from activity. Don't trade directionally.

### Tier 2: Market Makers
- Citadel Securities, Virtu, Jane Street
- Provide liquidity. Profit from spread and flow.
- Don't want directional risk (usually).
- MASSIVE information advantage (see all orders).

### Tier 3: Institutions
- Hedge funds, mutual funds, pension funds
- Move large sizes slowly (don't want to move price)
- Use dark pools, algos, VWAP strategies
- Have research, resources, time horizon

### Tier 4: Active Traders
- Prop shops, quant funds, day traders
- Trying to find edges, exploit inefficiencies
- Some sophisticated, many not
- We're in this tier (lower end)

### Tier 5: Passive/Retail
- Index funds, 401k investors, Robinhood traders
- Provide liquidity to everyone else
- Usually the exit liquidity
- Emotional, predictable, exploitable

## What I DON'T know:

- How do these tiers actually INTERACT?
- When does Tier 5 (retail) actually move markets?
- How do we avoid being exit liquidity?
- Can we act MORE like Tier 3/4 with Tier 5 resources?
